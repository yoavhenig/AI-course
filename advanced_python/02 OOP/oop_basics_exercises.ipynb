* Define the class `Person` with the following attributes and methods:
    * Attributes (in addition to `__init__()` arguments)
        * age (calculated)
        * is_alive – True
        * year_of_death – None
        * is_married – False
        * spouse – None
        * children – []
    * Methods
        * `__init__(self, name, id_number, gender, year_of_birth, current_year=2019)`
        * `grow(self, n)` – adds `n` to age
        * `got_married(self, spouse_name)` – updates the attributes spouse and is_married
        * `begat(self, children)` – updates the attribute children. Note that `children` may be either string or list.

* Create an instance of the class and test all the methods. Try to simulate the life story of someone you know. Print relevant attributes after every method, so that you’ll monitor the results.

```
class Person:
    def __init__(self, name, id_number, gender, year_of_birth, current_year=2019):
        pass
```

## Solution

```
class Person:
    def __init__(self, name, id_number, gender, year_of_birth, current_year=2019):
        self.name = name
        self.id_number = id_number
        self.gender = gender
        self.year_of_birth = year_of_birth
        self.age = current_year - self.year_of_birth
        
        self.is_alive = True
        self.year_of_death = None
        self.is_married = False
        self.spouse = None
        self.children = []
        
    def grow(self, n):
        if self.is_alive:
            self.age += n
        else:
            print("grass")
        
    def got_married(self, spouse_name):
        self.spouse = spouse_name
        self.is_married = True
        
    def begat(self, children):
        if isinstance(children, str):
            self.children.append(children)
        else:
            self.children.extend(children)
```

# Exercise 2

* Define the class `Queue` which implement the [Queue FIFO (First-In-First-Out) data structure](https://en.wikipedia.org/wiki/Queue_(abstract_data_type)). Its attributes and methods are:
    * Attribute
        * `queue`– the content of the queue (list)
    * Methods
        * `is_empty(self)` – returns True if the queue is empty 
        * `enqueue(self, element)` – pushes element at the “end” of queue
        * `dequeue(self)` – pops the “first” element from queue (if queue is not empty)
        * `get_len(self)` – returns the number of elements in queue

> **Note:** I chose arbitrarily to treat the queue as propagating from left to right, which means that new elements are inserted at the 0-th index of the list `queue`. You may flip this convention, but the final answer will be a little different.

## Solution

```
class Queue:
    def __init__(self):
        self.queue = []
        
    def is_empty(self):
        return len(self.queue) == 0
    
    def enqueue(self, element):
        self.queue.insert(0, element)
        
    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop()
    
    def get_len(self):
        return len(self.queue)
```
