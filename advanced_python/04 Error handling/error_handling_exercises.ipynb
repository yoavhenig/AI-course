# Exercise 1


In this exercise you will test the data compatibility of the content of the folder “exercise1”.
* The file “file_names.txt” contains a list of other file names, the content of which you should test.
* Every row in each of the other files contains a division exercise. A “legitimate” exercise is of the form “x,y,z” where x, y, and z are numbers.
* Your task is to go through all the lines in all the files and print whether or not x÷y=z.
* The code must check all the lines, and inform about any exception.
